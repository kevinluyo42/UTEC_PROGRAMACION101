```python
import pandas as pd

df_diabetes = pd.read_csv('diabetes.csv')
```

# Pregunta 1

```python
df_diabetes.head()
```

```python
# Número de filas y columnas
n_filas, n_columnas = df_diabetes.shape
print(f"{n_filas} filas")
print(f"{n_columnas} columnas")

# Tipo de dato de cada columna
print("\nFormatos:")
print(df_diabetes.dtypes)
```

```python
# Convertir "Outcome" a formato Category
df_diabetes['Outcome'] = df_diabetes['Outcome'].astype('category')

# Verificacion
df_diabetes.info()
```

# Pregunta 2

```python
df_filtrado = df_diabetes[df_diabetes['Glucose'] > 120]
```

```python
def clasificar_bmi(bmi):
    if bmi < 18.5:
        return 'Bajo'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Sobrepeso'
    else:
        return 'Obeso'

df_filtrado['BMI_Category'] = df_filtrado['BMI'].apply(clasificar_bmi)
```

```python
df_filtrado.head()
```

# Pregunta 3

```python
df_age_groups = pd.read_csv('age_groups.csv')
```

```python
#Aplicar left merge
df_combinado = df_diabetes.merge(df_age_groups, on='Age', how='left')
```

```python
df_combinado.head()
```

# Pregunta 4

```python
df_group = df_combinado.groupby('AgeGroup')
```

```python
promedios = df_group[['Glucose', 'BMI']].mean()
```

```python
promedios = promedios.sort_values(by='Glucose', ascending=False)
```

```python
promedios
```

# Pregunta 5

```python
df_filter2 = df_combinado[(df_combinado['AgeGroup'] == 'Adulto') & (df_combinado['Outcome'] == 1)]
```

```python
df_filter3 = df_filter2.nlargest(5, 'BMI')
```

```python
df_filter3[['Age', 'Glucose', 'BMI', 'DiabetesPedigreeFunction']]
```

# Pregunta 6

```python
pivot = df_combinado.pivot_table(
    values=['Glucose', 'BMI'],
    index='AgeGroup',
    columns='Outcome',
    aggfunc='mean'
)
```

```python
pivot
```

```python
filter = pivot['BMI'] > 30
pivot_filter = pivot[filter.any(axis=1)]
```
